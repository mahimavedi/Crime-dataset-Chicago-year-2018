```r
crimes <- read.csv("/Users/harigutta/Desktop/Crimes_-_2018.csv", stringsAsFactors = FALSE)
crimes_C <- na.omit(crimes)
head(crimes_C)
```

```r
crimes_C$Primary.Type <- as.factor(crimes_C$Primary.Type) 
crimes_C$Description <- as.factor(crimes_C$Description) 
crimes_C$Location.Description <- as.factor(crimes_C$Location.Description) 
crimes_C$IUCR <- as.factor(crimes_C$IUCR) 
 
crimes_C$Arrest[which(crimes_C$Arrest == "True")] <- 1
crimes_C$Arrest[which(crimes_C$Arrest == "False")] <- 0
 
crimes_C$Domestic[which(crimes_C$Domestic == "True")] <- 1
crimes_C$Domestic[which(crimes_C$Domestic == "False")] <- 0
summary(crimes_C)
```

```r
 library(ggplot2)
 primary_type <- ggplot(crimes_C, aes(Primary.Type))
 primary_type + geom_histogram(stat = "count") + coord_flip()
```

```r
install.packages("plotrix")
library(plotrix)
arrests <- table(crimes_C$Arrest)
bls <- paste(names(arrests), "\n", arrests, sep="")
pie3D(arrests, labels = lbls, 
    main="Arrests results (1 = True, 0 = False) from Crimes commited ")
domestic <- table(crimes_C$Domestic)
lbls <- paste(names(domestic), "\n", domestic, sep= "")
pie(domestic, labels = lbls, 
    main ="Domestic results (1 = True, 0 = False) for Crimes commited ")
top10_iucr <- tail(names(sort(table(crimes_C$IUCR))), 10)
```

```r
iucr_raw <- table(crimes_C$IUCR)
barplot(iucr_raw[order(iucr_raw, decreasing = TRUE)], xlim = c(0,11))
top10_description <- tail(names(sort(table(crimes_C$Description))), 10)
head(top10_description)
```
